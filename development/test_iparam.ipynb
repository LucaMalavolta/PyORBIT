```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib widget
```

```python
Rsun = 696000.#  Sun radius in km
d2s = 86400 
n_sample = 1000000
```

```python
rot_period = np.random.normal(7.5, 0.5, n_sample)
r_star =  np.random.normal(1.0, 0.02, n_sample)
v_eq =  2* np.pi * r_star * Rsun / (rot_period * d2s )
i_star = np.random.normal(98, 15, n_sample)
veq_sini = v_eq * np.sin(i_star / 180. * np.pi)
```

```python
plt.figure(figsize=(6,4))
plt.hist(v_eq, bins=50)
plt.show()

```

```python
ratio = veq_sini / v_eq
plt.figure(figsize=(6,4))
plt.hist(ratio, bins=50)
plt.show()
```

```python
ratio = veq_sini / v_eq
sel = (ratio > 1.)
ratio[sel] -= 2. 
i_star_derived = np.arcsin(ratio) / np.pi * 180.
print(np.sum(sel))
```

```python
plt.figure(figsize=(6,4))
plt.hist(i_star, bins=50)
plt.hist(i_star_derived, bins=50)
plt.show()
```
